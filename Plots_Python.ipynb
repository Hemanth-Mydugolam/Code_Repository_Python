## Plot Type - Best Use Case
- Line Plot - Time series, continuous data
- Bar Plot - Comparing categorical data
- Histogram - Distribution of numerical data
- Box Plot - Distribution comparison, outlier detection
- Scatter Plot - Relationship between two continuous variables
- Pie Chart - Proportions of a whole
- Heatmap - Data intensity, correlation matrix
- Pair Plot - Visualizing pairwise relationships
- Violin Plot - Data distribution across categories
- Bubble Chart - Visualizing three variables

#### 1. Line Plot
**Use Case:** To show trends over time or continuous data.<br>
**Why Best:** It clearly displays data points and their changes over a sequence, making it ideal for time series or any continuous data.<br>
**Example:** Stock prices over time, sales over a period of time

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example Data
data = {'Day': [1, 2, 3, 4, 5], 'Price': [10, 11, 12, 11, 13]}
df = pd.DataFrame(data)

# Plot
plt.plot(df['Day'], df['Price'], marker='o')
plt.title("Stock Prices Over Time")
plt.xlabel("Day")
plt.ylabel("Price")
plt.show()
```

### 2. Bar Plot
**Use Case:** To compare quantities across different categories.<br>
**Why Best:** It’s useful for categorical data as it shows clear, distinct comparisons between groups.<br>
**Example:** Sales across different product categories

```python
# Data
products = ['A', 'B', 'C']
sales = [100, 150, 90]

# Plot
plt.bar(products, sales)
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

### 3. Histogram
**Use Case:** To show the distribution of a numerical variable.<br>
**Why Best:** It's ideal for understanding the underlying distribution (e.g., normal, skewed) and identifying patterns like skewness or kurtosis.<br>
**Example:** Age distribution of a population.

```python
import numpy as np

# Example Data
ages = np.random.normal(30, 5, 100)

# Plot
plt.hist(ages, bins=10, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### 4. Box Plot (or Box-and-Whisker Plot)
**Use Case:** To summarize the distribution of a dataset and detect outliers. <br>
**Why Best:** It provides insights into the median, quartiles, and potential outliers, making it great for comparing distributions across categories. <br>
**Example:** Distribution of salaries in different departments

```python
# Example Data
data = {'IT': [60, 65, 70, 75, 80], 'HR': [50, 55, 60, 62, 63], 'Sales': [45, 55, 65, 70, 72]}
df = pd.DataFrame(data)

# Plot
df.boxplot()
plt.title("Salary Distribution by Department")
plt.ylabel("Salary (in 1000s)")
plt.show()
```

### 5. Scatter Plot
**Use Case:** To show the relationship between two variables.<br>
**Why Best:** It’s perfect for identifying correlations or patterns, especially in continuous data.<br>
**Example:** Relationship between advertising budget and sales revenue.

```python
# Example Data
budget = [10, 15, 20, 25, 30]
sales = [100, 120, 150, 170, 200]

# Plot
plt.scatter(budget, sales)
plt.title("Advertising Budget vs Sales")
plt.xlabel("Budget (in $1000s)")
plt.ylabel("Sales (in $1000s)")
plt.show()
```

### 6. Pie Chart
**Use Case:** To represent parts of a whole.<br>
**Why Best:** It’s best for showing proportional or percentage-based relationships, though it should be used sparingly due to potential misinterpretation.<br>
**Example:** Market share of different companies.

```python
# Data
labels = ['Company A', 'Company B', 'Company C']
sizes = [40, 35, 25]

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Market Share")
plt.show()
```

### 7. Heatmap
**Use Case:** To visualize data intensity or correlation between multiple variables.<br>
**Why Best:** It allows you to identify patterns and correlations visually, making it great for correlation matrices and large datasets.<br>
**Example:** Correlation matrix between features in a dataset.

```python
import seaborn as sns
import numpy as np

# Example Data
data = np.random.rand(5, 5)

# Plot
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

### 8. Pair Plot
**Use Case:** To visualize pairwise relationships between variables in a dataset.<br>
**Why Best:** It's useful for initial exploratory data analysis to understand the relationships between features.<br>
**Example:** Exploring the relationship between features in the Iris dataset.

```python
from sklearn.datasets import load_iris
import seaborn as sns

# Load Iris Dataset
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# Plot
sns.pairplot(df_iris)
plt.show()
```

### 9. Violin Plot
**Use Case:** To visualize the distribution of the data and its density across categories.<br>
**Why Best:** It combines the benefits of a box plot and kernel density plot, making it useful for showing distribution while highlighting any multimodal data.<br>
**Example:** Distribution of test scores across different classes

```python
# Example Data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Plot
plt.violinplot(data)
plt.title("Distribution of Scores by Class")
plt.show()
```

### 10. Bubble Chart
**Use Case:** To show the relationship between three variables.<br>
**Why Best:** It allows visualizing a third variable through the size of the bubbles, offering more information in a single plot.<br>
**Example:** Sales revenue (x), profit (y), and market share (bubble size).

```python
# Example Data
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]
size = [100, 300, 500, 700, 900]

# Plot
plt.scatter(x, y, s=size, alpha=0.5)
plt.title("Bubble Chart: Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

### 11. Area Plot
**Use Case:** To represent the cumulative totals over time.<br>
**Why Best:** It’s effective for visualizing parts of a whole and cumulative change over time, ideal for stacked data.<br>
**Example:** Cumulative revenue over different quarters for multiple products.

```python
# Example Data
df_area = pd.DataFrame({'Product A': [10, 20, 30, 40], 'Product B': [5, 15, 25, 35]})

# Plot
df_area.plot(kind='area')
plt.title("Cumulative Revenue Over Quarters")
plt.xlabel("Quarter")
plt.ylabel("Revenue")
plt.show()
```

### 12. Stacked Bar Plot
**Use Case:** To compare the composition of different categories while still showing totals.<br>
**Why Best:** It helps to compare parts of a whole while maintaining individual category breakdowns.<br>
**Example:** Sales by product across different regions.

```python
# Example Data
df_stacked = pd.DataFrame({'Region A': [10, 15, 20], 'Region B': [20, 25, 30]}, index=['Product A', 'Product B', 'Product C'])

# Plot
df_stacked.plot(kind='bar', stacked=True)
plt.title("Sales by Product Across Regions")
plt.ylabel("Sales")
plt.show()
```
